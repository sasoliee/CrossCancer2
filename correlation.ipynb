```python
import pandas as pd

# Step 1: Reshape `gene2` into long format for faster lookup
gene_long = gene2.melt(id_vars=gene2.columns[0], var_name="col", value_name="Gene")
gene_long = gene_long.drop(columns=["col"]).dropna()  # Drop unnecessary column
gene_long = gene_long.rename(columns={gene_long.columns[0]: "ID_REF"})  # Ensure first column is 'PI'

# Step 2: Filter only the relevant PIs for the given genes
gene_filtered = gene_long[gene_long["Gene"].isin(geneList)]

# Step 3: Merge once instead of filtering repeatedly
merged_df = gene_filtered.merge(reducedSM, on='ID_REF', how="left")
# Step 4: Compute mean efficiently using groupby()
final_geneAvg = merged_df.drop(columns=["ID_REF"]).groupby("Gene").mean()
# Display the final DataFrame
# final_geneAvg.iloc[:, row_list]
final_geneAvg
final_geneAvg.to_csv('Gene-Average.csv')

```
