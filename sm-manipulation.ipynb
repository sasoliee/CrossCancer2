```python
co = pd.read_csv("CO2.csv")

# Initialize subgroup lists
sb_one = []
sb_two = []
sb_three = []
sb_four = []
# sb_breast = [] # not to sure i think this section is unnecessary
# sb_lung = []
# sb_colon = []
# sb_kidney = []
# sb_uterus = []
# sb_ovary = []

# Iterate over rows
for index, row in co.iterrows():
    cancer_type = row['0']
    pid = row['Unnamed: 0']  # Assuming 'Unnamed: 0' is the patient/sample ID
    value = row[1]           # Assuming the second column has relevant values

    if cancer_type == 'Breast':
        sb_one.append([pid, value])
        sb_two.append([pid, value])
        sb_three.append([pid, value])
        sb_breast.append([pid, value])

    if cancer_type == 'Lung':
        sb_one.append([pid, value])
        sb_three.append([pid, value])
        sb_lung.append([pid,value])

    if cancer_type == 'Kidney':
        sb_two.append([pid, value])
        sb_kidney.append([pid,value])

    if cancer_type == 'Colon':
        sb_three.append([pid, value])
        sb_four.append([pid, value])
        sb_colon.append([pid, value])

    if cancer_type == 'Uterus':
        sb_two.append([pid, value])
        sb_four.append([pid, value])  
        sb_uterus.append([pid, value])

    if cancer_type == 'Ovary':
        sb_two.append([pid, value])
        sb_four.append([pid, value])
        sb_ovary.append([pid, value])


# Convert lists to DataFrames
subgroup1 = pd.DataFrame(sb_one, columns=['pid', 'Cancer'])
subgroup2 = pd.DataFrame(sb_two, columns=['pid', 'Cancer'])
subgroup3 = pd.DataFrame(sb_three, columns=['pid', 'Cancer'])
subgroup4 = pd.DataFrame(sb_four, columns=['pid', 'Cancer'])
subgroupB = pd.DataFrame(sb_breast, columns=['pid','Cancer'])
subgroupL = pd.DataFrame(sb_lung, columns=['pid', 'Cancer'])
subgroupC = pd.DataFrame(sb_colon, columns=['pid','Cancer'])
subgroupK = pd.DataFrame(sb_kidney, columns=['pid','Cancer'])
subgroupU = pd.DataFrame(sb_uterus, columns=['pid','Cancer'])
subgroupO = pd.DataFrame(sb_ovary, columns=['pid','Cancer'])

# Save to CSV
# subgroup1.to_csv('cancerOrder-S1.csv', index=False)
# subgroup2.to_csv('cancerOrder-S2.csv', index=False)
# subgroup3.to_csv('cancerOrder-S3.csv', index=False)
# subgroup4.to_csv('cancerOrder-S4.csv', index=False)

subgroupB.to_csv('subgroups/cancerOrder-Breast.csv', index=False)
subgroupL.to_csv('subgroups/cancerOrder-Lung.csv', index=False)
subgroupC.to_csv('subgroups/cancerOrder-Colon.csv', index=False)
subgroupK.to_csv('subgroups/cancerOrder-Kidney.csv', index=False)
subgroupU.to_csv('subgroups/cancerOrder-Uterus.csv', index=False)
subgroupO.to_csv('subgroups/cancerOrder-Ovary.csv', index=False)
```
